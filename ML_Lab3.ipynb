```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

data = pd.read_csv("https://raw.githubusercontent.com/dhruvpathak1/ml_dataset/main/online_shoppers_intention.csv")
```

```python
from sklearn import preprocessing 
#before normalisation
print(data['ExitRates'].plot(kind = 'bar'))
```

```python
# copy the data
df_min_max_scaled = data.copy()
# apply normalization techniques to column 'ExitRates'
column = 'ExitRates'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min()) 
# view normalized data
print(df_min_max_scaled['ExitRates'].plot(kind='bar'))
```

```python
# applying label encoding to the 'BounceRates' column 
le = preprocessing.LabelEncoder()
data['BounceRates'] = le.fit_transform(data['BounceRates'])
# as you can see the data has been encoded 
data['BounceRates'].head()
```

```python
# to check if there is skewness in features 
# skewness along the index axis
print(data.skew(axis = 0, skipna = True))

data.ProductRelated_Duration.hist()
plt.title('Skewness in ProductRelated_Duration')
```

```python
data.OperatingSystems.hist()
plt.title('Skewness in Operating Systems')
```

```python
#function to return plots for the feature
#importing necessary libraries
import scipy.stats as stats
import pylab
import seaborn as sns
def normality(data,feature):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.kdeplot(data[feature])
    plt.subplot(1,2,2)
    stats.probplot(data[feature],plot=pylab)
    plt.show()

#performing logarithmic transformation on the feature
data['ExitRates']=np.log(data['ExitRates'])
#plotting to check the transformation
normality(data,'ExitRates')

```

```python
# using feature engineering wherever necessary 
# normalising all data with large values and are of float datatype

#performing logarithmic transformation on the feature
data['Administrative_Duration']=np.log(data['Administrative_Duration'])
#plotting to check the transformation
normality(data,'Administrative_Duration')
```

```python
#performing logarithmic transformation on the feature
data['TrafficType']=np.log(data['TrafficType'])
#plotting to check the transformation
normality(data,'TrafficType')
```
