```python
import pandas as pd
from matplotlib import pyplot as plt

df2 = pd.read_csv("https://raw.githubusercontent.com/dhruvpathak1/ml_dataset/main/online_shoppers_intention.csv", 
                  usecols=["ProductRelated", "BounceRates", "ExitRates", "ProductRelated_Duration","VisitorType"])
plt.scatter(df2['BounceRates'], df2['VisitorType'])
plt.xlabel('Bounce Rates')
plt.ylabel('Visitor Type')
plt.title("Bounce Rates v/s Visitor Type")
plt.show()
```

```python
plt.figure(figsize = (4, 4))
plt.bar(df2['BounceRates'], df2['ProductRelated_Duration'])
plt.xlabel('Bounce Rates')
plt.ylabel('VisitorType')
plt.title("Bounce Rates v/s VisitorType")
plt.show()
```

```python
plt.figure(figsize = (4, 4))
plt.hist(df2['BounceRates'], bins = 15)
plt.title("Itching v/s Visitor Type")
plt.xlabel('BounceRates')
plt.ylabel('Visitor Type')
plt.show()
```
