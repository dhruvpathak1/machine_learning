### **K Means**

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

x = df2.iloc[:, [3,4]].values
wcss=[]
for i in range(1,11): 
     kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10,random_state=0 )

kmeans.fit(x)
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(df2.iloc[:,[3,4]])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,5))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="orange", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

As demonstrated in the graph above, errors drop exponentially as the K Mean value goes up. 

```python
kmeans = KMeans(n_clusters=5, init ='k-means++', max_iter=300, n_init=10,random_state=0 )
y_kmeans = kmeans.fit_predict(x)
print(kmeans.inertia_)
```

```python
plt.scatter(x[y_kmeans==0, 0], x[y_kmeans==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(x[y_kmeans==1, 0], x[y_kmeans==1, 1], s=100, c='yellow', label ='Cluster 2')
plt.scatter(x[y_kmeans==2, 0], x[y_kmeans==2, 1], s=100, c='blue', label ='Cluster 3')
plt.scatter(x[y_kmeans==3, 0], x[y_kmeans==3, 1], s=100, c='plum', label ='Cluster 4')
plt.scatter(x[y_kmeans==4, 0], x[y_kmeans==4, 1], s=100, c='orange', label ='Cluster 5')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='purple', label = 'Centroids')
plt.title('Clusters')
plt.xlabel('Spending Score')
plt.ylabel('Income')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


x = df2.iloc[:, [2,3]].values
std_slc = StandardScaler()
X_std = std_slc.fit_transform(x)
clt = DBSCAN()

model = clt.fit(X_std)

clusters = pd.DataFrame(model.fit_predict(X_std))
df2["Cluster"] = clusters

fig = plt.figure(figsize=(10,10)); ax = fig.add_subplot(111)
scatter = ax.scatter(df2['Age'],df2['Annual Income (k$)'], c=df2["Cluster"],s=50)
ax.set_title("DBSCAN Clustering")
ax.set_xlabel("Age")
ax.set_ylabel("Annual Income")
plt.colorbar(scatter)
plt.show()
```

Finding No. of Clusters


```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize

data=df2.iloc[:, [2,3]].values
db=DBSCAN(eps=3,min_samples=5,metric='euclidean') 
#eps:maximum distance between two samples 
#min_sample:umber of samples (or total weight) in a neighborhood 
#metric:metric use to calculate distance
model=db.fit(data)
label=model.labels_
#identifying the points which makes up our core points
sample_cores=np.zeros_like(label,dtype=bool)

sample_cores[db.core_sample_indices_]=True

#Calculating the number of clusters

n_clusters=len(set(label))- (1 if -1 in label else 0)
print('No of clusters:',n_clusters)
```

```python
y_means = db.fit_predict(data)
plt.figure(figsize=(7,5))
plt.scatter(data[y_means == 0, 0], data[y_means == 0, 1], s = 50, c = 'pink')
plt.scatter(data[y_means == 1, 0], data[y_means == 1, 1], s = 50, c = 'yellow')
plt.scatter(data[y_means == 2, 0], data[y_means == 2, 1], s = 50, c = 'cyan')
plt.scatter(data[y_means == 3, 0], data[y_means == 3, 1], s = 50, c = 'magenta')
plt.scatter(data[y_means == 4, 0], data[y_means == 4, 1], s = 50, c = 'orange')
plt.scatter(data[y_means == 5, 0], data[y_means == 5, 1], s = 50, c = 'blue')
plt.scatter(data[y_means == 6, 0], data[y_means == 6, 1], s = 50, c = 'red')
plt.scatter(data[y_means == 7, 0], data[y_means == 7, 1], s = 50, c = 'black')
plt.scatter(data[y_means == 8, 0], data[y_means == 8, 1], s = 50, c = 'violet')
plt.xlabel('Annual Income in (1k)')
plt.ylabel('Spending Score from 1-100')
plt.title('Clusters of data')
plt.show()
```

Various algorithms produce different numbers of clusters. Since it creates clusters based on epsilon, min points, and noise, DBscan is a more efficient method than K-means and Hierarchical clustering. It forecasts the core points, boundary points, and outliers independently and effectively.
